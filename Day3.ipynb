Ranking Rows of Pandas DataFrame

```python
movies = {'Name': ['The Godfather', 'Bird Box', 'Fight Club'],
         'Year': ['1972', '2018', '1999'],
         'Rating': ['9.2', '6.8', '8.8']}
```

```python
df=pd.DataFrame(movies)
df
```

```python
df['Rating_Rank']=df['Rating'].rank(ascending=1)
```

```python
df = df.set_index('Rating_Rank')
print(df)
```

```python
df=df.sort_index()
```

```python
df
```

Q)take an example of marks scored by 4 students. We will rank the students based on the highest mark they have scored

```python
student_details = {'Name':['Raj', 'Raj', 'Raj', 'Aravind', 'Aravind', 'Aravind',
                             'John', 'John', 'John', 'Arjun', 'Arjun', 'Arjun'],
                   'Subject':['Maths', 'Physics', 'Chemistry', 'Maths', 'Physics',
                             'Chemistry', 'Maths', 'Physics', 'Chemistry', 'Maths',
                                                           'Physics', 'Chemistry'],
                   'Marks':[80, 90, 75, 60, 40, 60, 80, 55, 100, 90, 75, 70]
               }
```

Select row with maximum and minimum value in Pandas dataframe

Q)Creating a Dataframe to select rows with max and min values in Dataframe


```python
dict1 = {'Driver': ['Hamilton', 'Vettel', 'Raikkonen',
                    'Verstappen', 'Bottas', 'Ricciardo',
                    'Hulkenberg', 'Perez', 'Magnussen',
                    'Sainz', 'Alonso', 'Ocon', 'Leclerc',
                    'Grosjean', 'Gasly', 'Vandoorne',
                    'Ericsson', 'Stroll', 'Hartley', 'Sirotkin'],
 
         'Points': [408, 320, 251, 249, 247, 170, 69, 62, 56,
                    53, 50, 49, 39, 37, 29, 12, 9, 6, 4, 1],
 
         'Age': [33, 31, 39, 21, 29, 29, 31, 28, 26, 24, 37,
                 22, 21, 32, 22, 26, 28, 20, 29, 23]}
```

```python
df=pd.DataFrame(dict1)
df
```

```python

df.head()
```

```python
#For max
print(df['Points'].max())

```

```python
print(df[df.Points == df.Points.max()]) #for who scored max points
```

```python
#for max age
df.Age.max()
```

```python
#max age of whom
df[df.Age== df.Age.max()]
```

Shows min on Driver, Points, Age columns

```python
#for points
df.Points.min()
```

```python
#min points of whom
df[df.Points== df.Points.min()]
```

```python
#for age
df.Age.min()
```

```python
#min age of whom
df[df.Age==df.Age.min()]
```

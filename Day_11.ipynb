```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Multiple Plots

```python
# Using the add_axes() method 
from matplotlib.figure import Figure
```

```python
fig= plt.figure(figsize=(5,4))
ax1= fig.add_axes([0.1,0.1,0.8,0.8])
ax2 = fig.add_axes([0.5,0.5,0.3,0.3])
ax1.plot([5, 4, 3, 2, 1], [2, 3, 4, 5, 6])
ax2.plot([1, 2, 3, 4, 5], [2, 3, 4, 5, 6])
plt.show()
```

```python
#Using subplot() method
x = [3, 1, 3]
y = [3, 2, 1]
z = [1, 3, 1]
plt.figure()
plt.subplot(121)
plt.plot(x,y)
plt.subplot(122)
plt.plot(z,y)
plt.show()
```

# Barchart

```python
#plt.bar(x, height, width, bottom, align)
x = [3, 1, 3, 12, 2, 4, 4]
y = [3, 2, 1, 4, 5, 6, 7]
plt.bar(x,y)
plt.title("Bar chart")
plt.show()
```

# Histograms

```python
x = [1, 2, 3, 4, 5, 6, 7, 4]
plt.hist(x,bins=[1,2,3,4,5,6,7])
plt.title("Histogram")
plt.legend("Mark")
plt.show()
```

# Scatter plot

```python
x = [3, 1, 3, 12, 2, 4, 4]
y = [3, 2, 1, 4, 5, 6, 7]
plt.scatter(x,y)
plt.legend("A")
plt.title("Scatterplot")
plt.show()
```

# Pie Chart

```python
x = [1, 2, 3, 4]
e  =(0.1, 0, 0, 0) #e= explode,
plt.pie(x, explode=e)
plt.title("Pie chart")
plt.show()
```

# 3D Plots 

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
z = [1, 8, 27, 64, 125]
fig= plt.figure()
ax= plt.axes(projection= '3d')
ax.plot3D(z, y, x)
```

# Change Fonts in matplotlib

```python
from scipy import signal
t = np.linspace(0,1,1000, endpoint=True)
plt.plot(t,signal.square(2*np.pi*5*t))
plt.xlabel("Time (Seconds)", fontname="Brush Script MT", fontsize=18)
plt.ylabel("Amplitude", fontname="Brush Script MT", fontsize=18)
plt.show()
```

```python
# Change the font just for the title
x = [1, 2, 3, 4, 5, 6]
y = [0, 2, 4, 6, 8, 10]
plt.scatter(x,y)
plt.title("Line Graph",fontname='Franklin Gothic Medium', fontsize=18)
```

```python
#Change the font just for the title and axis labels.
x = [1, 2, 3, 4, 5, 6]
y = [0, 2, 4, 6, 8, 10]
plt.plot(x,y)
plt.xlabel("Years", fontname="Gabriola", fontsize=18)
plt.ylabel("Population (million)", fontname="Gabriola", fontsize=18)
plt.title("Line Graph - Geeksforgeeks",
           fontname='Franklin Gothic Medium', fontsize=18)
plt.show()
```

# Set Plot Background Color in Matplotlib

```python
#Default color plot
student_marks = [50, 60, 70, 80, 90]
student_grade = ['B', 'B', 'B+', 'B+', 'A']
plt.plot(student_marks,student_grade)
plt.xlabel("student_marks", fontweight='bold')
plt.ylabel("student_grade", fontweight='bold')
plt.title("Student Marks v/s Student Grade")
plt.show()
```

```python
# Setting background color to yellow
student_marks = [50, 60, 70, 80, 90]
student_grade = ['B', 'B', 'B+', 'B+', 'A']
plt.plot(student_marks, student_grade)
plt.xlabel("student_marks", fontweight='bold')
ax = plt.axes()
ax.set_facecolor("yellow")
plt.ylabel("student_grade", fontweight='bold')
plt.title("Student Marks v/s Student Grade")
plt.show()
```

# Add Text Inside the Plot in Matplotlib

```python
#Adding mathematical equations inside the plot
x = np.arange(-10, 10, 0.01)
y = x**2
plt.text(-5, 60, 'Parabola $Y = x^2$', fontsize = 22)
plt.plot(x, y, c='g')
plt.xlabel("X-axis", fontsize = 15)
plt.ylabel("Y-axis",fontsize = 15)
plt.show()
```

```python
#Adding rectangular box around the text by using the keyword bbox. bbox is a dictionary of Rectangle properties.
x = np.arange(-10, 10, 0.01)
y = x**2
plt.xlabel("X-axis", fontsize = 15)
plt.ylabel("Y-axis",fontsize = 15)
plt.text(-5, 60, 'Parabola $Y = x^2$', fontsize = 22, 
         bbox = dict(facecolor = 'red', alpha = 0.5))
plt.plot(x, y, c = 'g')
plt.show()
```

```python
#Adding the text Sine wave inside the plot.
x = np.arange(0, 10, 0.1)
y = np.sin(x)
plt.plot(x,y)
plt.text(3.5, 0.9, 'Sine wave', fontsize = 23)
plt.xlabel('X-axis', fontsize = 15)
plt.ylabel('Y-axis', fontsize = 15)
plt.show()
```
